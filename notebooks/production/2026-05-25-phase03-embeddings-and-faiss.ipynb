# Phase 3 — Embeddings & FAISS

**Goal:** Turn chunks into dense vectors, store them in a FAISS index alongside their metadata, and retrieve the top-k nearest chunks for a query.

By the end you will have:

1. An embedding for every chunk produced in Phase 2.
2. A persisted FAISS index (`data/vector_store/`) you can reload later.
3. A `VectorStore` class with the methods we need for `src/mrta/retrieval/vector_store.py`: `add`, `search`, `save`, `load`.
4. A first qualitative test: ask "what is multi-head attention?" and look at what comes back.


## 3.1 — Load chunks from Phase 2


````python
import os, sys
from pathlib import Path

repo_root = Path.cwd()
while not (repo_root / "pyproject.toml").is_file() and repo_root != repo_root.parent:
    repo_root = repo_root.parent
os.chdir(repo_root)
sys.path.insert(0, str(repo_root))

from mrta.core.schemas import Chunk

chunks_path = Path("data/processed/chunks.jsonl")
chunks = [Chunk.model_validate_json(line) for line in chunks_path.read_text().splitlines()]
print(f"Loaded {len(chunks)} chunks")
print("First:", chunks[0].chunk_id, "-", chunks[0].text[:80].replace("\n", " "), "...")
````

## 3.2 — Pick an embedder

We default to `sentence-transformers/all-MiniLM-L6-v2`:
- 22M parameters, runs on CPU at ~1k sentences/sec on a modern laptop;
- 384-dim output;
- trained on diverse semantic pairs (good general-purpose retrieval).

For higher quality at the cost of speed, swap in `BAAI/bge-large-en-v1.5` (1024-dim, top of the MTEB English leaderboard).


````python
import numpy as np
from mrta.retrieval.embedder import Embedder

embedder = Embedder()  # reads settings.embedding_model
DIM = embedder.dim
print(f"Embedder    : {embedder.model_name}")
print(f"Output dim  : {DIM}")
````

## 3.3 — Embed all chunks

We normalize embeddings (`normalize_embeddings=True`) so we can use **inner-product** as the FAISS metric, which is mathematically equivalent to cosine similarity for unit-norm vectors.


````python
import time

texts = [c.text for c in chunks]
t0 = time.time()
embeddings = embedder.embed(texts)   # ← mrta's method
# embeddings = embedder.encode(
#     texts,
#     batch_size=32,
#     show_progress_bar=True,
#     normalize_embeddings=True,
#     convert_to_numpy=True,
# ).astype("float32")
dt = time.time() - t0
print(f"\nEmbedded {len(texts)} chunks in {dt:.1f}s ({len(texts)/dt:.0f}/s)")
print(f"Shape: {embeddings.shape}")
print(f"Norms (should all be ~1.0): min={np.linalg.norm(embeddings, axis=1).min():.3f}, "
      f"max={np.linalg.norm(embeddings, axis=1).max():.3f}")
````

## 3.4 — Build a FAISS index

For paper-scale corpora (<1M vectors) `IndexFlatIP` is the right choice — exact nearest-neighbor, no training, instant.

For larger corpora you would switch to `IndexHNSWFlat` or `IndexIVFPQ`. We will demo the swap below.


````python
import faiss

index = faiss.IndexFlatIP(DIM)
index.add(embeddings)
print(f"FAISS index built. ntotal = {index.ntotal}, dim = {index.d}")
````

## 3.5 — A query, end-to-end


````python
query = "What is multi-head attention?"
# q_vec = embedder.encode([query], normalize_embeddings=True).astype("float32")
q_vec = embedder.embed([query])

scores, indices = index.search(q_vec, k=5)
print(f"Query: {query}\n")
for rank, (i, s) in enumerate(zip(indices[0], scores[0]), start=1):
    c = chunks[i]
    print(f"[{rank}] score={s:.3f}  page={c.page}  chunk={c.chunk_id}")
    print("    ", c.text[:160].replace("\n", " "), "...\n")
````

## 3.6 — Wrap it all in a reusable `VectorStore`

This is the class that lives in `src/mrta/retrieval/vector_store.py`. Two key responsibilities:

1. Maintain a **parallel array** of metadata so we can map FAISS row indices back to chunks.
2. Provide `save` / `load` so we do not re-embed on every restart.


````python
from mrta.retrieval.vector_store import VectorStore

# Build, save, reload, re-test.
store = VectorStore(embedder)
store.add(chunks)
store.save("data/vector_store/aiayn")

reloaded = VectorStore.load("data/vector_store/aiayn", embedder)
hits = reloaded.search("How does scaled dot-product attention work?", k=3)
for h in hits:
    print(f"  page {h.page}  -- {h.text[:90]}...")
````

## 3.7 — Index variants for scale

| Index type           | When to use                                         | Pros                            | Cons                        |
|----------------------|-----------------------------------------------------|---------------------------------|-----------------------------|
| `IndexFlatIP`        | < 1M vectors, exact nearest neighbor                | Simplest, exact, no training    | O(n) per query              |
| `IndexHNSWFlat`      | 1M–10M vectors, fast queries                        | Sub-millisecond queries         | Build time, RAM-heavy       |
| `IndexIVFPQ`         | 10M–100M+, RAM-constrained                          | Compressed storage              | Approximate, needs training |
| Qdrant (server)      | Production: filtering, persistence, multi-tenancy   | Metadata filters, REST API      | Extra service to run        |

## Exercise 2 — Add doc_id filter to VectorStore.search

FAISS doesn't support metadata filtering natively — it only returns indices. The filter is applied after retrieval: fetch more candidates than needed, then keep only those matching the doc_id.


def search(self, query: str, k: int = 5, doc_id: str | None = None) -> list[Chunk]:
    if not self._chunks:
        return []
    q = self._embedder.embed([query])
    # Fetch more candidates when filtering so we still return k results
    fetch_k = k * 5 if doc_id else k
    scores, idx = self._ensure_index().search(q, min(fetch_k, len(self._chunks)))
    results = [self._chunks[i] for i in idx[0] if 0 <= i < len(self._chunks)]
    if doc_id:
        results = [c for c in results if c.doc_id == doc_id]
    return results[:k]
The k * 5 over-fetch compensates for chunks that get filtered out. If you have 10 chunks from paper A and 1000 from paper B, and you filter on paper A, the top-5 FAISS results might all be from paper B — you need to reach further into the index.

Usage:


vs.search("What is attention?", k=5, doc_id="attention_paper_3f8a2c")

## Exercise 3 — Tiny eval set, measure recall@5

Recall@5 = "is the answer page in the top 5 retrieved chunks?"


from mrta.evaluation.metrics import recall_at_k

# 10 question / expected-page pairs (build these by hand from your PDF)
eval_set = [
    {"question": "What is the attention mechanism?",         "expected_page": 1},
    {"question": "What dataset was used for training?",      "expected_page": 6},
    {"question": "How many encoder layers does the model use?", "expected_page": 3},
    {"question": "What is the BLEU score on WMT 2014?",     "expected_page": 8},
    {"question": "What optimizer was used?",                 "expected_page": 6},
    {"question": "What is multi-head attention?",            "expected_page": 3},
    {"question": "How is positional encoding computed?",     "expected_page": 4},
    {"question": "What is the role of the feed-forward layer?", "expected_page": 4},
    {"question": "How large is the vocabulary?",             "expected_page": 6},
    {"question": "What are the limitations mentioned?",      "expected_page": 9},
]

hits_at_5 = []
for item in eval_set:
    results = vs.search(item["question"], k=5)
    retrieved_pages = [str(c.page) for c in results]
    expected = [str(item["expected_page"])]
    score = recall_at_k(retrieved_pages, expected, k=5)
    hits_at_5.append(score)
    status = "✓" if score == 1.0 else "✗"
    print(f"{status} [{score:.0f}] {item['question'][:55]}")

print(f"\nRecall@5: {sum(hits_at_5)/len(hits_at_5):.2f}  ({sum(hits_at_5):.0f}/{len(hits_at_5)} correct)")
recall_at_k is already in mrta.evaluation.metrics — it checks what fraction of expected_docs appear in retrieved_sources[:k]. Here each question has one expected page, so the score is binary: 1.0 if the right page was retrieved, 0.0 if not.
